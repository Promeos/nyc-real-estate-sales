```python
# Rename columns
df.columns = df.columns.str.replace('\n', '')
df.columns = df.columns.str.lower()
df.columns = [column.replace(' ', '_') for column in df.columns]
```

```python
df.drop(columns = ['tax_class_as_of_final_roll_18/19',
                   'ease-ment',
                   'building_class_as_of_final_roll_18/19',
                   'apartment_number',
                   'residential_units',
                   'commercial_units'],
       inplace=True)
```

```python
df = df.dropna()
df = df.
```

```python
sns.set_context('talk')
df.hist(figsize=(16, 16), bins=30);
plt.tight_layout()
```

```python
nonnumeric_data = df.select_dtypes(exclude='number')

nonnumeric_data.columns
```

```python
sns.set_context('talk')
plt.figure(figsize=(10, 10))
sns.countplot(data=nonnumeric_data, y=f'neighborhood')
plt.tight_layout();
```
