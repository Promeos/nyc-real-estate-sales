```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Rename columns for easier data manipulation
df.columns = df.columns.str.replace('\n', '')
df.columns = df.columns.str.lower()
df.columns = [column.replace(' ', '_') for column in df.columns]
```

```python
df.borough.value_counts()
```

```python
df.drop(columns = ['tax_class_as_of_final_roll_18/19',
                   'ease-ment',
                   'building_class_as_of_final_roll_18/19',
                   'apartment_number',
                   'residential_units',
                   'commercial_units',
                   'borough'],
       inplace=True)

df = df.dropna()
```

```python
sns.set_context('talk')
df.hist(figsize=(16, 16), bins=30);
plt.tight_layout()
```

```python
numeric_columns = df.select_dtypes('number').columns.to_list()
```

```python
sns.set_context('talk')
sns.pairplot(df[numeric_columns[:5]])
```

```python
sns.set_context('talk')
sns.pairplot(df[numeric_columns[5:]])
```
